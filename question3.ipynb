```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def save_maxfig(fig,fig_name):
    fig.set_size_inches(5, 5)
    plt.savefig(fig_name,dpi=300)
```

# A

```python
def original_case(c):
    x = np.array([80,50])
    y = np.array([30,60])
    plt.scatter(x,y,color='cornflowerblue',label='Original case')
    plt.text(x[0]+1,y[0],'Dell')
    plt.text(x[1]+1,y[1],'Lenovo')
    plt.text(c[0]-3,c[1],'HP')
    plt.xlabel('Quality',fontsize=14)
    plt.ylabel('Work duration',fontsize=14)

fig1 = plt.figure('1')
c = [77,27]
original_case(c)
plt.title('Decoy effect',fontsize=16)
plt.scatter(c[0],c[1],color='lightcoral',label='Added option')
save_maxfig(fig1,'3fig1')
fig2 = plt.figure('2')
c = [100,5]
original_case(c)
plt.title('Compromise effect',fontsize=16)
plt.scatter(c[0],c[1],color='lightcoral',label='Added option')
save_maxfig(fig2,'3fig2')
```
